```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

# --- Configuración de Rutas ---
# Define la ruta base (donde están tus archivos CSV)
DATASET_DIR = Path("/home/lauramartinez/my_env/QPU147-monitoring/datasets")
# Si la carpeta ya existe y contiene los CSV, esta línea es solo una verificación:
DATASET_DIR.mkdir(parents=True, exist_ok=True) 

# ✅ Archivos de entrada: Ahora se especifican con la extensión .csv
csv_file_names = [
    "20251002_Q0_coherence_monitoring_log.csv",
    "20251003_Q0_coherence_monitoring_log.csv",
    "20251004_1_Q0_coherence_monitoring_log.csv",
    "20251004_2_Q0_coherence_monitoring_log.csv",
    "20251005_Q0_coherence_monitoring_log.csv",
    "20251006_1_Q0_coherence_monitoring_log.csv",
    "20251006_2_Q0_coherence_monitoring_log.csv",
    "20251006_3_Q0_coherence_monitoring_log.csv",
    "20251006_4_Q0_coherence_monitoring_log.csv",
    # "20251007_Q0_coherence_monitoring_log.csv"
]

# Rutas completas a los archivos de origen
csv_paths = [DATASET_DIR / name for name in csv_file_names]

# --- Lectura y Combinación de archivos CSV ---
DATA_DIR = DATASET_DIR
OUTPUT_CSV = Path("merged_coherence_data.csv")
GAP_THRESHOLD_MINUTES = 15

# Usamos directamente las rutas definidas
csv_files = csv_paths 

if not csv_files:
    raise FileNotFoundError(f"No se encontraron archivos CSV en la carpeta: {DATA_DIR}. Verifique que los nombres de los archivos son correctos.")

dfs = [] 
for f in csv_files: 
    # ✅ CLAVE: Leer como CSV y forzar la columna 'Timestamp' a ser cadena (str)
    # Esto evita el error de notación científica al convertir a datetime.
    df = pd.read_csv(f, dtype={'Timestamp': str}) 
    dfs.append(df) 
    print(f"✅ Archivo leído: {f.name}")

# Combina todos los dataframes
data = pd.concat(dfs, ignore_index=True)

# -----------------------------------------------------------------------------
## Limpieza y Procesamiento

# Conversión de timestamps: Ahora la conversión debe funcionar correctamente
data["Timestamp"] = pd.to_datetime(data["Timestamp"])
data = data.sort_values("Timestamp").reset_index(drop=True)

# Detección de gaps
data["TimeDiff"] = data["Timestamp"].diff().dt.total_seconds() 
gap_threshold = GAP_THRESHOLD_MINUTES * 60 
data["Gap"] = data["TimeDiff"] > gap_threshold 

# Guarda el conjunto de datos combinado en el directorio de datos
output_csv_path = DATA_DIR / OUTPUT_CSV
data.to_csv(output_csv_path, index=False)
print(f"\nDatos combinados guardados en {output_csv_path}")

# -----------------------------------------------------------------------------
## Gráfico de Tiempos de Coherencia

plt.figure(figsize=(10, 5))
plt.plot(data["Timestamp"], data["T1"] * 1e6, marker='o', linestyle='-', label='T1 (μs)')
plt.plot(data["Timestamp"], data["T2*"] * 1e6, marker='x', linestyle='--', label='T2* (μs)')
plt.plot(data["Timestamp"], data["T2E"] * 1e6, marker='s', linestyle=':', label='T2E (μs)')

# Marca visualmente las brechas (gaps)
for idx, row in data.iterrows(): 
    if row["Gap"]: 
        plt.axvspan(data["Timestamp"].iloc[idx-1], row["Timestamp"], color='gray', alpha=0.2) 

plt.title("Coherence Time Monitoring")
plt.xlabel("Timestamp")
plt.ylabel("Coherence Times (μs)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
